```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
# loading the dataset
df = pd.read_csv("C:/Users/matth/Downloads/House Price India.csv")
df.head()
```

```python
df.info()
```

### Univariate Analysis

```python
sns.histplot(df["Built Year"])
```

```python
sns.histplot(df["number of bedrooms"])
```

```python
sns.distplot(df["number of floors"])
```

### Bi - Variate Analysis

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Lattitude', y='Longitude')
plt.title('Scatter Plot ')
plt.show()
```

### Multi-Variate Analysis

```python
corr_mat = df.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr_mat, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
sns.pairplot(df[['number of bedrooms', 'number of views', 'Price']])
plt.title('Pair Plot')
plt.show()
```
